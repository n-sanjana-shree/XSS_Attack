```python
import pandas as pd
from imblearn.over_sampling import ADASYN

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/XSS_dataset2_engineered.csv')

# Separate the features and the target variable
X = df.drop(columns=['Class'])
y = df['Class']

# Define the ADASYN oversampling strategy
adasyn = ADASYN(sampling_strategy='minority', random_state=42)

# Apply the ADASYN oversampling strategy
X_resampled, y_resampled = adasyn.fit_resample(X, y)

# Create a dataframe from the resampled features and target variable
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['Class'] = y_resampled

# Shuffle the resampled data
df_resampled = df_resampled.sample(frac=1, random_state=42)

# Save the resampled data to a new CSV file
df_resampled.to_csv('resampled_dataset.csv', index=False)
```

```python
def remove_collinear_features(x, threshold):
   

    # Calculate the correlation matrix
    corr_matrix = x.corr()
    iters = range(len(corr_matrix.columns) - 1)
    drop_cols = []

    # Iterate through the correlation matrix and compare correlations
    for i in iters:
        for j in range(i+1):
            item = corr_matrix.iloc[j:(j+1), (i+1):(i+2)]
            col = item.columns
            row = item.index
            val = abs(item.values)

            # If correlation exceeds the threshold
            if val >= threshold:
                # Print the correlated features and the correlation value
                print(col.values[0], "|", row.values[0], "|", round(val[0][0], 2))
                drop_cols.append(col.values[0])

    # Drop one of each pair of correlated columns
    drops = set(drop_cols)
    x = x.drop(columns=drops)

    return x
```
